```python
import qiskit.quantum_info as qi
from qiskit.circuit.library import FourierChecking
from qiskit.visualization import plot_histogram
```

```python
f=[1,-1,-1,-1]
g=[1,1,-1,-1]
```

```python
circuit = FourierChecking(f=f, g=g)
circuit.draw(output='mpl')
```

```python
zeroVector = qi.Statevector.from_label('00')
stateVector = zeroVector.evolve(circuit)
probabilities = stateVector.probabilities_dict()
plot_histogram(probabilities)
#00 probability identifiers Fourier comparability of f & g
```
